```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Automobile.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# distribution of continuous variable using histogram
cont_var = ['mpg','cylinders','acceleration']
for var in cont_var:
    sns.histplot(df[var])
    plt.show()
```

```python
#identify the relationship b/w 2 continuous variables using scatter plot
sns.scatterplot(data=df,x='displacement',y='acceleration')
```

```python
#find and display the frequency of categorical values using countplot
sns.countplot(data=df,y='cylinders')
```

```python
#apply pointplot to display one continuous and one categorical variable
sns.pointplot(data=df,x='cylinders',y='acceleration')
```
