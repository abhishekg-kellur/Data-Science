```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("basket.csv")
```

```python
df_new = pd.DataFrame(df)
```

```python
df_dum = pd.get_dummies(df_new)
```

```python
frequent_items = apriori(df_dum,min_support=0.03,use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items,metric='confidence',min_threshold=0.1)
rules
```
